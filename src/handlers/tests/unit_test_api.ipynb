```python
date_start='2022-11-11'
date_end='2022-12-12'
keyword='meal kit'
```

```python
def process():
    if all_articles['status'] != "ok":
        raise ValueError("something wrong with request params")
    else:
        inital_df=process_page(all_articles) #breaks json down and reads into df
        if int(all_articles['totalResults']) >100:
            total_pages=int(int(all_articles['totalResults']) / 100)
            print(f"total pages are {total_pages}")
            for i in range(2,total_pages+1):
                print(f"on page {i}")
                json= newsapi.get_everything(q=keyword,from_param=date_start,to=date_end,language='en',sort_by='relevancy',page=i)
                next_df=process_page(json)
                print(f"len of next df : {len(next_df)}")
                inital_df=pd.concat((inital_df, next_df), axis = 0)
                print(f"here is len of new df {len(inital_df)}")
            return inital_df
        
```

```python
df=process()
```

```python
def process_page(all_articles):
    p_list=[]
    for i in range(len(all_articles['articles'])):
        source=all_articles['articles'][i]['source']['name'] or "NA"
        source_id=all_articles['articles'][i]['source']['id'] or "NA"
        author=all_articles['articles'][i]['author'] or "NA"
        title=all_articles['articles'][i]['title'] or "NA"
        description=all_articles['articles'][i]['description'] or "NA"
        url=all_articles['articles'][i]['url'] or "NA"
        url_img=all_articles['articles'][i]['urlToImage'] or "NA"
        date=all_articles['articles'][i]['publishedAt'] or "NA"
        content=all_articles['articles'][i]['content'] or "NA"
        p_list.append([source,source_id,author,title,description,url,url_img, date, content])
        df=pd.DataFrame(p_list, columns=['Source', 'Source_ID','Author','Title','Description','Url','Url_Img','Date_Published','Content'])
    return df
```

```python
df=process_page(all_articles)
```
